```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the dataset
df = pd.read_csv("C:/Users/priya/Desktop/diabetes predication/diabetes.csv")

# Define features (X) and target (y)
X = df.iloc[:,0:8].astype(float)  # All columns except 'Outcome'
y = df.iloc[:,8]              # The 'Outcome' column (target)

# Now split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=7)
```

**🚀 Define the Model**


```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Define the model
model = Sequential()

# Input layer + first hidden layer
model.add(Dense(12, input_dim=8, activation='relu'))

# Second hidden layer
model.add(Dense(8, activation='relu'))

# Output layer
model.add(Dense(1, activation='sigmoid'))  # Sigmoid for binary classification
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### 🧠 Fit the Model and Make Predictions

- **Epoch** – One complete pass through all of the rows in the training dataset.
- **Batch** – One or more samples considered by the model within an epoch before weights are updated.


```python
# fit the keras model on the dataset
model.fit(X_train, y_train, epochs=150, batch_size=10)
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=10, validation_split=0.2)
```

```python
# prediction of the keras model
y_pred = model.predict(X_test)
```

**🔍 Evaluate the Model**


```python
# evaluate the keras model
accuracy = model.evaluate(X_test, y_test)
print('Accuracy: ' % (accuracy*100))
```
